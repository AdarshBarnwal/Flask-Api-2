```python
Q1. Explain GET and POST methods.
Ans:-

Get method:- GET method is used to retrieve information from the server.   
             Actually It is the most common method which can be used to send data in the unencrypted form to the server.
         (Through URL)
             
Post method:- POST method is used to submit data to the server to create or update a resource.
             Actually It is used to send the form data to the server. The server does not cache the data transmitted using the post method.
         (Through BODY)
```

```python
Q2. Why is request used in Flask?
Ans:- To access the incoming data in flask that's ehy request used in flask and the request object holds all incoming data from the request,
      which includes the mimetype, referrer, ip address, raw data, http method, and headers, among other things.
      
        in the client-server architecture, the request object contains all the data that is sent from the client to the server.
```

```python
Q3. Why is redirect() used in Flask?
Ans:- Redirect() used in flask to redirect users to a specified url and assign a specified status code. when this function is called,
      a responce object is returned, and the redirection happens to the target location with the status code.
```
